# Task 1 : Import the dataset

Import the dataset into a `pandas` dataframe.


```python
df = pd.read_csv(file_name, header=None)
print(df.head(10))
```

Add the headers to the dataframe, as mentioned in the project scenario. 


```python
headers = ["age", "gender", "bmi", "no_of_children", "smoker", "region", "charges"]
df.columns = headers
```

Now, replacing the '?' entries with 'NaN' values.


```python
df.replace('?', np.nan, inplace = True)
```

# Task 2 : Data Wrangling


Using `dataframe.info()` identify the columns that have some 'Null' (or NaN) information.


```python
print(df.info())
```

Handle missing data:

- For continuous attributes (e.g., age), replace missing values with the mean.
- For categorical attributes (e.g., smoker), replace missing values with the most frequent value.
- Update the data types of the respective columns.
- Verify the update using `df.info()`.


```python
is_smoker = df['smoker'].value_counts().idxmax()
df["smoker"].replace(np.nan, is_smoker, inplace=True)

# age is a continuous variable, replace with mean age
mean_age = df['age'].astype('float').mean(axis=0)
df["age"].replace(np.nan, mean_age, inplace=True)

# Update data types
df[["age","smoker"]] = df[["age","smoker"]].astype("int")

print(df.info())
```

Also note, that the `charges` column has values which are more than 2 decimal places long. Updating the `charges` column such that all values are rounded to nearest 2 decimal places. Verify conversion by printing the first 5 values of the updated dataframe.


```python
df[["charges"]] = np.round(df[["charges"]],2)
print(df.head())
```

# Task 4 : Model Development

Fitting a linear regression model that may be used to predict the `charges` value, just by using the `smoker` attribute of the dataset. Print the $ R^2 $ score of this model.


```python
X = df[['smoker']]
Y = df['charges']
lm = LinearRegression()
lm.fit(X,Y)
print(lm.score(X, Y))
```

Fitting a linear regression model that may be used to predict the `charges` value, just by using all other attributes of the dataset. Print the $ R^2 $ score of this model. You should see an improvement in the performance.


```python
Z = df[["age", "gender", "bmi", "no_of_children", "smoker", "region"]]
lm.fit(Z,Y)
print(lm.score(Z, Y))
```

Creating a training pipeline that uses `StandardScaler()`, `PolynomialFeatures()` and `LinearRegression()` to create a model that can predict the `charges` value using all the other attributes of the dataset. This shows even further improvement in the performance.


```python
Input=[('scale',StandardScaler()), ('polynomial', PolynomialFeatures(include_bias=False)), ('model', LinearRegression())]
pipe=Pipeline(Input)
Z = Z.astype(float)
pipe.fit(Z,Y)
ypipe=pipe.predict(Z)
print(r2_score(Y,ypipe))
```

# Task 5 : Model Refinement



```python
x_train, x_test, y_train, y_test = train_test_split(Z, Y, test_size=0.2, random_state=1)
```

Initializing a Ridge regressor that uses hyperparameter $ \alpha = 0.1 $. Fit the model using training data data subset. Print the $ R^2 $ score for the testing data.


```python
RidgeModel=Ridge(alpha=0.1)
RidgeModel.fit(x_train, y_train)
yhat = RidgeModel.predict(x_test)
print(r2_score(y_test,yhat))
```

```python
print('predicted:', yhat[0:4])
print('test set :', y_test[0:4].values)
```

To check which value of alpha gives the minimum error, we need to use different hyperparameters and then check their score. Sklearn has the class GridSearchCV to make the process of finding the best hyperparameter simpler.

```python
from sklearn.model_selection import GridSearchCV
```

We create a dictionary of parameter values.

```python
parameters= [{'alpha': [0.001,0.1,1, 10, 100, 1000, 10000, 100000, 100000]}]
parameters
```

Create a Ridge regression object:

```python
RR=Ridge()
RR
```

Create a ridge grid search object:

```python
Grid1 = GridSearchCV(RR, parameters,cv=4)
```

Fit the model:

```python
Grid1.fit(Z, Y)
```

```python
BestRR=Grid1.best_estimator_
BestRR
```

```python
BestRR.score(x_test, y_test)
```

This shows that our assumption of alpha = 0.1 estimates the best results.

Apply polynomial transformation to the training parameters with degree=2. Use this transformed feature set to fit the same regression model, as above, using the training subset. Print the $ R^2 $ score for the testing subset.


```python
pr = PolynomialFeatures(degree=2)
x_train_pr = pr.fit_transform(x_train)
x_test_pr = pr.fit_transform(x_test)
RidgeModel.fit(x_train_pr, y_train)
y_hat = RidgeModel.predict(x_test_pr)
print(r2_score(y_test,y_hat))
```
